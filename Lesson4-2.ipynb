```python
import csv
with open(file='f:\students.txt',encoding='utf-8') as file:
    reader = csv.reader(file)
    #next(reader) 跳過第一行行
    students:list = list(reader)

for student in students:
    if student[0] == '姓名:student1':
        print(student)
```
