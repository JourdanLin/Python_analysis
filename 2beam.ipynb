```python
N = 44  # Total number of offsets
p = 33  # Number of dice per offset
spindles_per_head = 7  # Number of spindles per head

# Split offsets between Head1 and Head2
head1_offsets = list(range(0, N // 2))  # Offset 0 to Offset 21
head2_offsets = list(range(N // 2, N))  # Offset 22 to Offset 43

# Track current state
current_head1_offset_idx = 0  # Index into head1_offsets
current_head2_offset_idx = 0  # Index into head2_offsets
current_die_head1 = 1  # Start at U1
current_die_head2 = 1  # Start at U1
round_number = 1

# Flags to track if heads are finished
head1_finished = False
head2_finished = False

while not (head1_finished and head2_finished):
    # Round for Head1
    if not head1_finished:
        print(f"第{round_number}輪")
        if current_head1_offset_idx < len(head1_offsets):
            current_offset = head1_offsets[current_head1_offset_idx]
            # Process up to 7 dice
            for s in range(spindles_per_head):
                if current_die_head1 <= p:
                    spindle = s + 1
                    die = f"U{current_die_head1}"
                    print(f"Head1 - Spindle{spindle} - Offset {current_offset} - {die}")
                    current_die_head1 += 1
            # Check if all dice for this offset are processed
            if current_die_head1 > p:
                current_head1_offset_idx += 1  # Move to next offset
                current_die_head1 = 1  # Reset to U1
        else:
            head1_finished = True
        round_number += 1

    # Round for Head2
    if not head2_finished:
        print(f"第{round_number}輪")
        if current_head2_offset_idx < len(head2_offsets):
            current_offset = head2_offsets[current_head2_offset_idx]
            # Process up to 7 dice
            for s in range(spindles_per_head):
                if current_die_head2 <= p:
                    spindle = s + 1
                    die = f"U{current_die_head2}"
                    print(f"Head2 - Spindle{spindle} - Offset {current_offset} - {die}")
                    current_die_head2 += 1
            # Check if all dice for this offset are processed
            if current_die_head2 > p:
                current_head2_offset_idx += 1  # Move to next offset
                current_die_head2 = 1  # Reset to U1
        else:
            head2_finished = True
        round_number += 1
```
